```python
import matplotlib.pyplot as p

p.plot(3,4)
p.show()
```

```python
x = [0,1.5,4.5,3,1.5,4.5,6]
y = [0,1.5,1.5,3,1.5,1.5,0]

p.plot(x,y,'m')
p.show()
```

```python
import numpy as np

x = np.linspace(0, 3* np.pi, 101)
y = np.sin(x)

p.plot(x,y,'m')
p.show()
```

```python
fig, ax = p.subplots(3,3)
for axes in ax.flatten():
  axes.plot(np.random.randn(4,4))

p.show()
```

```python
fig, ax = p.subplots()

x = np.linspace(0,10,100)

for i in np.linspace(0,1,50):
  ax.plot(x,x**i,color=[i/2,0,i])

p.show()
```
